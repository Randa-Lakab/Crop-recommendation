# 🌾 Crop Recommendation — Exploratory Data Analysis
A structured EDA of the crop recommendation dataset before model training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os
sys.path.insert(0, os.path.join('..', 'model'))
from train import generate_synthetic_data, FEATURES

sns.set_theme(style='whitegrid', palette='Set2')
%matplotlib inline
plt.rcParams['figure.dpi'] = 120
```

## 1. Load Data

```python
DATA = os.path.join('..', 'data', 'crop_recommendation.csv')
df = pd.read_csv(DATA) if os.path.exists(DATA) else generate_synthetic_data()
print(f'Shape: {df.shape}')
df.head()
```

## 2. Basic Statistics

```python
df.describe().round(2)
```

## 3. Class Distribution

```python
fig, ax = plt.subplots(figsize=(12, 5))
counts = df['label'].value_counts()
sns.barplot(x=counts.index, y=counts.values, palette='YlGn', ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.set_title('Samples per Crop', fontsize=14, fontweight='bold')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4. Feature Distributions

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for i, feat in enumerate(FEATURES):
    ax = axes[i // 4][i % 4]
    df[feat].hist(bins=40, ax=ax, color='#2ECC71', edgecolor='white')
    ax.set_title(feat)
    ax.set_ylabel('Count')
fig.delaxes(axes[1][3])
fig.suptitle('Feature Distributions', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 5. Correlation Heatmap

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(df[FEATURES].corr(), annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, linewidths=0.5, ax=ax)
ax.set_title('Feature Correlation Matrix', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 6. Feature Box-plots by Crop (Top 8)

```python
top8 = df['label'].value_counts().head(8).index
sub  = df[df['label'].isin(top8)]

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for i, feat in enumerate(FEATURES):
    ax = axes[i // 4][i % 4]
    sub.boxplot(column=feat, by='label', ax=ax, rot=45,
                boxprops=dict(color='#1A6B2F'),
                whiskerprops=dict(color='#1A6B2F'),
                medianprops=dict(color='#F39C12', linewidth=2))
    ax.set_title(feat); ax.set_xlabel('')
    plt.sca(ax); plt.xticks(fontsize=7)
fig.delaxes(axes[1][3])
fig.suptitle('Feature Distribution per Crop (Top 8)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
